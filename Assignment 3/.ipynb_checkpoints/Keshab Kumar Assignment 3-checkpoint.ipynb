## Visualizations using Python

```python
plt.figure(figsize=(5, 5))
plt.pie(df['Gender'].value_counts(), labels= df['Gender'].unique(), autopct = '%1.1f%%', startangle = 180, colors = ['green', 'orange'])
plt.title('Gender Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 5))
plt.pie(df['Marital Status'].value_counts(), labels= df['Marital Status'].unique(), autopct = '%1.1f%%', startangle = 180, colors = ["lightblue", "lightgreen", "lightcoral"])
plt.title('Marital Status Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 6))
plt.bar(df["Location"].unique(), df["Location"].value_counts())
for location, value in zip(df['Location'].unique(), df['Location'].value_counts()):
  plt.text(location, value, str(value), ha='center',  va='bottom')
plt.xlabel("Location")
plt.ylabel("Frequency")
plt.title("Location Frequency Distribution")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.figure(figsize=(5, 6))
plt.bar(df["Cuisines"].unique(), df["Cuisines"].value_counts())
for cuisine, value in zip(df['Cuisines'].unique(), df['Cuisines'].value_counts()):
  plt.text(cuisine, value, str(value), ha='center',  va='bottom')
plt.xlabel("Cuisines")
plt.ylabel("Frequency")
plt.title("Cuisines Frequency Distribution")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
avg_overall_rating = df.groupby(['Cuisines', 'Gender'])['Overall Rating'].mean().unstack()
avg_overall_rating
```

```python
plt.figure(figsize=(5, 6))
avg_overall_rating.plot(kind='bar')

plt.title("Average Overall Rating for Cuisines by Gender")
plt.xlabel("Cuisines")
plt.ylabel("Average Overall Rating")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
avg_budget_rating = df.groupby(['Activity', 'Gender'])['Overall Rating'].mean().unstack()
avg_budget_rating
```

```python
plt.figure(figsize=(5, 6))
avg_budget_rating.plot(kind='bar')

plt.title("Average Budget Rating for Activity by Gender")
plt.xlabel("Activity")
plt.ylabel("Average Budget Rating")
plt.xticks(rotation=45, ha="right")
plt.show()
```
